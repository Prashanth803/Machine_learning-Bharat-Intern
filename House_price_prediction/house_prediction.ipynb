```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
y_predictreg=reg.predict(x_test)
#not bad can we make it still better
#yes by using GradientBoosting Regressor
```

```python
from sklearn import ensemble 
```

```python
clf=ensemble.GradientBoostingRegressor(n_estimators=400,max_depth=5,min_samples_split=2,learning_rate=0.1,loss='ls')
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
#as u can see the accuracy has increased to ~90%
y_predict=clf.predict(x_test)
```

```python
print(y_predict,y_predictreg)
# calculating the differce in regression model and booster regression model
print((y_predictreg-y_predict))
```
